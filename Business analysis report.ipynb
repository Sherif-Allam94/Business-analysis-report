# Growth plan (20%, 30%, and 40%)

```python
#20%

df3= pd.read_excel("1.xlsx", sheet_name="Sheet8")
df3.set_index('index')
df3=df3.transpose()

new_header = df3.iloc[0] #grab the first row for the header
df3 = df3[1:] #take the data less the header row
df3.columns = new_header #set the header row as the df header

df3=df3.reset_index()

df3.info()
#df3.to_excel('20.xlsx')
```

```python
#30%

df4= pd.read_excel("1.xlsx", sheet_name="Sheet9")

df4.set_index('index')

df4=df4.transpose()

new_header = df4.iloc[0] #grab the first row for the header
df4 = df4[1:] #take the data less the header row
df4.columns = new_header #set the header row as the df header

df4=df4.reset_index()

df4.info()

df4.head()

#df4.to_excel('30.xlsx')
```

```python
#40%

df5= pd.read_excel("1.xlsx", sheet_name="Sheet10")

df5.set_index('index')

df5=df5.transpose()

new_header = df5.iloc[0] #grab the first row for the header
df5 = df5[1:] #take the data less the header row
df5.columns = new_header #set the header row as the df header

df5=df5.reset_index()

df5.info()

#df5.to_excel('40.xlsx')
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='index', y='TPV', data=df3, label="TPV @20")
sns.lineplot(x='index', y='TPV', data=df4, label="TPV @30")
sns.lineplot(x='index', y='TPV', data=df5, label="TPV @40")
plt.ylabel("TPV", fontsize=15)
plt.xlabel('date', fontsize=15)
plt.title("TPV mom @ 20,30,40%", fontsize=18)
plt.yticks(fontsize=15)
plt.xticks(fontsize=14,rotation=45)
plt.legend(fontsize=15);
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='index', y='Monthly Revenue', data=df3, label="Monthly Revenue @20")
sns.lineplot(x='index', y='Monthly Revenue', data=df4, label="Monthly Revenue @30")
sns.lineplot(x='index', y='Monthly Revenue', data=df5, label="Monthly Revenue @40")
plt.ylabel("Monthly Revenue", fontsize=15)
plt.xlabel('date', fontsize=15)
plt.title("Monthly Revenue @ 20,30,40%", fontsize=18)
plt.yticks(fontsize=15)
plt.xticks(fontsize=14,rotation=45)
plt.legend(fontsize=15)
plt.savefig("mrev.jpg", dpi=500)
plt.show();
```
